```python
"""
=============================================================================================================================================================================================================
    üßÆ CALCULADORA INTERATIVA EM PYTHON
=============================================================================================================================================================================================================

üìå Descri√ß√£o:
Este programa implementa uma calculadora interativa com opera√ß√µes matem√°ticas b√°sicas e c√°lculo de IRPF (Imposto de Renda Pessoa F√≠sica). Tamb√©m mant√©m um hist√≥rico das opera√ß√µes realizadas.

üë®‚Äçüíª Autor: CLEITON HENTGES
üìÖ Data: [27/08/2025]
üíª Linguagem: Python 3.x
"""
```

```python
import os

historico = []

def limpar_tela():
    os.system('cls' if os.name == 'nt' else 'clear')

def mostrar_historico():
    print("\n===== HIST√ìRICO DE OPERA√á√ïES =====")
    if not historico:
        print("Nenhuma opera√ß√£o foi realizada ainda.")
    else:
        for item in historico:
            print(item)
    print("==================================\n")

def calcular_irpf(salario):
    if salario <= 2112.00:
        aliquota = 0
        deducao = 0
    elif salario <= 2826.65:
        aliquota = 0.075
        deducao = 158.40
    elif salario <= 3751.05:
        aliquota = 0.15
        deducao = 370.40
    elif salario <= 4664.68:
        aliquota = 0.225
        deducao = 651.73
    else:
        aliquota = 0.275
        deducao = 884.96

    imposto_mensal = (salario * aliquota) - deducao if aliquota > 0 else 0
    salario_liquido_mensal = salario - imposto_mensal

    imposto_anual = imposto_mensal * 12
    salario_liquido_anual = salario_liquido_mensal * 12

    return aliquota, imposto_mensal, salario_liquido_mensal, imposto_anual, salario_liquido_anual

# Loop principal
while True:
    print("=== CALCULADORA INTERATIVA ===\n")
    print("Escolha a opera√ß√£o que deseja realizar:")
    print("1 - Adi√ß√£o")
    print("2 - Subtra√ß√£o")
    print("3 - Multiplica√ß√£o")
    print("4 - Divis√£o")
    print("5 - Ver hist√≥rico")
    print("6 - Sair")
    print("7 - Calcular Imposto de Renda (IRPF)\n")

    opcao = input("Digite o n√∫mero da op√ß√£o desejada: ").strip()

    if opcao == "1":
        print("\n[ADI√á√ÉO]")
        a = float(input("Parcela 1: ").replace(',', '.'))
        b = float(input("Parcela 2: ").replace(',', '.'))
        resultado = a + b
        operacao_str = f"{a} + {b} = {resultado}"
        print("Resultado:", operacao_str)
        historico.append(operacao_str)

    elif opcao == "2":
        print("\n[SUBTRA√á√ÉO]")
        a = float(input("Minuendo: ").replace(',', '.'))
        b = float(input("Subtraendo: ").replace(',', '.'))
        resultado = a - b
        operacao_str = f"{a} - {b} = {resultado}"
        print("Resultado:", operacao_str)
        historico.append(operacao_str)

    elif opcao == "3":
        print("\n[MULTIPLICA√á√ÉO]")
        a = float(input("Fator 1: ").replace(',', '.'))
        b = float(input("Fator 2: ").replace(',', '.'))
        resultado = a * b
        operacao_str = f"{a} * {b} = {resultado}"
        print("Resultado:", operacao_str)
        historico.append(operacao_str)

    elif opcao == "4":
        print("\n[DIVIS√ÉO]")
        a = float(input("Dividendo: ").replace(',', '.'))
        b = float(input("Divisor: ").replace(',', '.'))
        if b == 0:
            print("Erro: Divis√£o por zero n√£o √© permitida.")
        else:
            resultado = a / b
            resto = a % b
            operacao_str = f"{a} / {b} = {resultado} | Resto: {resto}"
            print("Resultado:", operacao_str)
            historico.append(operacao_str)

    elif opcao == "5":
        mostrar_historico()

    elif opcao == "6":
        print("\nSaindo da calculadora... At√© logo!")
        break

    elif opcao == "7":
        print("\n[C√ÅLCULO DE IMPOSTO DE RENDA]")
        salario = float(input("Digite o sal√°rio bruto mensal (R$): ").replace(',', '.'))
        aliquota, imposto_mensal, liquido_mensal, imposto_anual, liquido_anual = calcular_irpf(salario)

        if imposto_mensal == 0:
            resultado = f"Sal√°rio Mensal: R$ {salario:.2f} ‚Üí Isento de IR"
        else:
            resultado = (f"[MENSAL] Bruto: R$ {salario:.2f} | "
                        f"Al√≠quota: {aliquota * 100:.1f}% | "
                        f"IRPF: R$ {imposto_mensal:.2f} | "
                        f"L√≠quido: R$ {liquido_mensal:.2f}\n"
                        f"[ANUAL] Bruto: R$ {salario*12:.2f} | "
                        f"IRPF: R$ {imposto_anual:.2f} | "
                        f"L√≠quido: R$ {liquido_anual:.2f}")
        print("Resultado:\n", resultado)
        historico.append("IRPF ‚Üí " + resultado)


    else:
        print("\nOp√ß√£o inv√°lida. Por favor, selecione uma op√ß√£o de 1 a 7.")

    input("\nPressione Enter para continuar...")
    limpar_tela()
    print("==================================\n")
```
